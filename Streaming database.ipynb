```python
# function for donating-------------------------------------------------------------------------
def donate():
    
    # makes 3 arrays for each of streamer, viewer and donation tables
    strea = np.array(pd.read_sql('select * from streamer',connection))
    view = np.array(pd.read_sql('select * from viewer',connection))
    dons = np.array(pd.read_sql('select * from donation',connection))
    
    # initializes array for streamer and viewer ids
    s_ida = []
    v_ida = []
    
    # populates the arrays with the respective ids
    for d in range(len(strea)):
        s_ida.append(strea[d][0])
    for d in range(len(view)):
        v_ida.append(view[d][0])
        
    # creates new id for the new donation
    newdon_id = 200 + len(dons)
    
    # takes input of viewer id and checks if it exists in viewer ids array
    curview = int(input('input your viewer id as 1#: '))
    if curview in v_ida:
        twallet = 0  # initializes wallet value
        
        # finds wallet value and stores it
        for n in range(len(view)):
            if curview == view[n][0]: # 0th index is always the v_id
                twallet = view[n][3]  # 3rd index is always the wallet value
                

        # takes input of streamer id and checks if it exists in streamer ids array
        curstreamer = int(input('input the streamer id as #: '))
        if curstreamer in s_ida:
            
            # takes input of wanted donation
            don = int(input('input your donation amount: '))
            # stores value of difference between viewer wallet and donation and checks if its positive
            remain = twallet - don
            if (remain) > 0:
                
                for n in range(len(strea)):
                    if curstreamer==strea[n][0]:  # 0th index is always the s_id and 2nd index is donation value
                        newdon = strea[n][2] + don  # stores new streamer donation value

                        # updates streamer donation value with new donation
                        cursor.execute(f"update streamer set donations={newdon} where s_id={curstreamer}")
                        # updates viewer wallet with remaining money
                        cursor.execute(f"update viewer set wallet={remain} where v_id={curview}")
                        # creates new donation record with values
                        cursor.execute(f"insert into donation values({newdon_id},{curstreamer}," +
                                       f"{curview},{don},'{date.today()}')")
                        
                        # prints updated and new records
                        print('records updated')
                        print('\n')
                        print(pd.read_sql(f'select * from streamer where s_id={curstreamer}',connection))
                        print('\n')
                        print(pd.read_sql(f'select * from viewer where v_id={curview}',connection))
                        print('\n')
                        print(pd.read_sql(f'select * from donation where don_id={newdon_id}',connection))
                        print('\n')

            else:
                print('not enough money in wallet \n') # output if not enough money in wallet
                    
            
        else:
            print('streamer does not exist \n') # output if streamer does not exist


            
    else:
        print('viewer does not exist please register \n')  # output if viewer does not exist

        
# function for showing tables-------------------------------------------------------------------------
def showtable():
    
    print('choose a table from streamer , viewer , game , donation , stream')
    table_select = str(input('input table name: ')) # input table name
    
    # if table exists shows it using panda read_sql
    try:
        print(pd.read_sql(f'select * from {table_select}',connection))
        print('\n')
        
    except:
        print('table does not exist \n')   # ouput if table name doesnt exist
        
        
# function for showing records-------------------------------------------------------------------------
def showrecord():
    
    print('choose a table from streamer , viewer , game , donation , stream')
    table_select = str(input('input table name: ')) # input table name
    
    if table_select == 'streamer':
        # asks for streamer id input and displays record
        s_id = int(input('input streamer id as #: '))
        print(pd.read_sql(f'select * from streamer where s_id={s_id}',connection))
        print('\n')
        
    elif table_select == 'viewer':
        # asks for viewer id input and displays record
        v_id = int(input('input viewer id as 1#: '))
        print(pd.read_sql(f'select * from viewer where v_id={v_id}',connection))
        print('\n')
        
    elif table_select == 'game':
        # asks for game id input and displays record
        g_id = int(input('input game id as 1##: '))
        print(pd.read_sql(f'select * from game where g_id={g_id}',connection))
        print('\n')
        
    elif table_select == 'donation':
        # asks for donation id input and displays record
        d_id = int(input('input donation id as 2##: '))
        print(pd.read_sql(f'select * from donation where don_id={d_id}',connection))
        print('\n')
        
    elif table_select == 'stream':
        # asks for stream id input and displays record
        st_id = int(input('input stream id as 3##: '))
        print(pd.read_sql(f'select * from stream where st_id={st_id}',connection))
        print('\n')
        
    else:
        print('table does not exist \n') # ouput if table name doesnt exist


# function for adding record-------------------------------------------------------------------------
def addrecord():
    print('choose a table from streamer , viewer , game , stream')
    table_select = str(input('input table name: ')) # input table name
    
    if table_select == 'streamer':
        # inputs for record parameters
        s_id = int(input('input streamer id as #: '))
        s_name = str(input('input streamer name: '))
        s_don = int(input('input streamer donation: '))
        # inserts record in table
        cursor.execute(f"insert into streamer values({s_id},'{s_name}',{s_don},'{date.today()}')")
        print('record added')
        # prints the record
        print(pd.read_sql(f'select * from streamer where s_id={s_id}',connection))
        print('\n')
        
    elif table_select == 'viewer':
        # inputs for record parameters
        v_id = int(input('input viewer id as 1#: '))
        v_name = str(input('input viewer name: '))
        v_wal = int(input('input viewer wallet: '))
        # inserts record in table
        cursor.execute(f"insert into viewer values({v_id},'{v_name}','{date.today()}',{v_wal})")
        print('record added')
        # prints the record
        print(pd.read_sql(f'select * from viewer where v_id={v_id}',connection))
        print('\n')
        
    elif table_select == 'game':
        # inputs for record parameters
        g_id = int(input('input game id as 1##: '))
        g_name = str(input('input game name: '))
        g_dev = str(input('input game developer: '))
        # inserts record in table
        cursor.execute(f"insert into game values({g_id},'{g_name}','{g_dev}')")
        print('record added')
        # prints the record
        print(pd.read_sql(f'select * from game where g_id={g_id}',connection))
        print('\n')
        
    elif table_select == 'stream':
        # inputs for record parameters
        st_id = int(input('input stream id as 3##: '))
        s_id = int(input('input streamer id as #: '))
        g_id = int(input('input game id as 1##: '))
        # inserts record in table
        cursor.execute(f"insert into stream values({st_id},{s_id},{g_id},'{date.today()}')")
        print('record added')
        # prints the record
        print(pd.read_sql(f'select * from stream where st_id={st_id}',connection))
        print('\n')
        
    else:
        print('table does not exist \n') # ouput if table name doesnt exist
    
# function for advance mysql commands-------------------------------------------------------------------------
def advancemysql():
    inp = str(input('input mysql command: ')) # command input
    print(inp + ' \n') # prints input
    cursor.execute(inp)  # executes input
    print(cursor.fetchall()) # prints command
```
